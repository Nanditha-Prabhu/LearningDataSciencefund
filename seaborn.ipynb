```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 700)
y = np.sin(10*x)
z = np.cos(10*x)
```

```python
sns.set_style("whitegrid")
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z)
plt.show()
```

```python
from numpy import sqrt


def cfunc(x, y):
    return sqrt(1000 - x**2 - y**2)
```

```python
a, b = np.linspace(-10, 10, 100), np.linspace(-10, 10, 100)
```

```python
a, b = np.meshgrid(a, b)
c = cfunc(a, b)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(a, b, c)
plt.show()
```
