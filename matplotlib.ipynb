MATPLOTLIB

```python
#import numpy to generate random numbers
import numpy as np
#import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate random numbers
randomNum = np.random.rand(10)
```

```python
#view them
print(randomNum)
```

```python
#select style of the plot
style.use('ggplot')
#plot random no
plt.plot(randomNum, 'b', label='first', linewidth=1) #data, color, legend, line-thickness
#x axis is number of random numbers
plt.xlabel('range')
#y axis is random number
plt.ylabel('numbers')
#title of plot
plt.title('GGPLOT')
#display
plt.legend()
plt.show()
```

```python
#creating dataset
ingredients = 'laung', 'cardamom', 'elaichi', 'bay-leaf', 'turmeric', 'star-anise'
percentile = [75, 70, 60, 80, 95, 55] 
```

```python
#set size of img
plt.figure(figsize=(3, 3))
#explode largest pie of the chart
explode=(0, 0, 0, 0, 0.3, 0)
#plot it in pie chart
plt.pie(percentile, labels=ingredients, explode=explode, autopct='%1.1f%%', startangle=90)
#set title
plt.title('SPICES')
#display
plt.show()
```
